### Заполнение пропусков в столбце `ceiling_height`

Изучим данные в столбце:

```python
pd.DataFrame(df['ceiling_height'].describe())
```

Диапазон значений высоты потолков составляет 1 и 100 м, что явно является ошибкой. Изучим распределение выбросов:

```python
df['ceiling_height'].hist(bins=50, range=(5,100))
plt.xlabel('Высота, м')
plt.ylabel('Количество')
plt.show()
```

Можно предположить, что в значениях в диапазоне от 20 до 40 была допущена ошибка в разрядности, 100 и значения ниже 20 являются артефактами. Напишем функцию для замены выбросов и пропусков на медианное значение высоты:

```python
print('Пропуски до:', df['ceiling_height'].isna().sum())

# объявим глобальные переменные
q1 = df['ceiling_height'].quantile(0.25)
q3 = df['ceiling_height'].quantile(0.75)    
iqr = 1.5 * (q3 - q1)
median = df['ceiling_height'].median()  

# напишем функцию для проверки вхождения в межквартильный размах
def ceiling_height_adj(height):  
    if height > (q1 - iqr) and height < (q3 + iqr):
        return height
    if height/10 > (q1 - iqr) and height/10 < (q3 + iqr):  # снижение разрядности 
        return height/10
    return median

# применим функцию к столбцу
df['ceiling_height'] = df['ceiling_height'].apply(ceiling_height_adj)

print('Пропуски после:', df['ceiling_height'].isna().sum())
```

Проверим распределение данных после исправления:

```python
pd.DataFrame(df['ceiling_height'].describe())
```

Выбросы и пропуски в столбце `ceiling_height` устранены.

Согласно действующему законодательству, учетная норма площади жилого помещения на одного человека в Санкт-Петербурге составляет
9 м² общей площади жилого помещения, значения площади ниже данного, вероятнее всего, являются артефактами. Заполним пропуски и значения ниже порогового медианным значением жилой площади по квартирам с аналогичным количеством комнат, но предварительно проанализируем столбец с количеством комнат:

```python
display(pd.DataFrame(df['rooms'].describe()))

df['rooms'].hist(bins=20)
plt.xlabel('Комнат')
plt.ylabel('Количество')
plt.xticks(np.arange(0, 19, 2))
plt.show()

print('Объявлений без указанного количества комнат:', df[df['rooms'] == 0]['rooms'].count())
```

В массиве 197 объявлений, где количество комнат не указано. Заменим 0 на медианное значение количества комнат по квартирам с аналогичным метражом:

```python
print('Объявлений без указанного количества комнат до:', df[df['rooms'] == 0]['rooms'].count())

valid_rooms_area_list = df[df['rooms'] > 0]['total_area'] # создаем список метражей, по которым есть данные по числу комнат

for area in df['total_area'].unique(): # для заполнения пропусков используем цикл
    if area in valid_rooms_area_list:
        median = round(df.loc[df['total_area'] == area, 'rooms'].median(), 0)  # определяем медиану для каждого метража
        df.loc[(df['rooms'] == 0) & (df['total_area'] == area), 'rooms'] = median  # используем метод loc для заполнения пропусков
    else:  #  если соответствующий метраж отсуствует, используем округление до десятков
        median = round(df.loc[round(df['total_area'], -1) == round(area, -1), 'rooms'].median(), 0) 
        df.loc[(df['rooms'] == 0) & (round(df['total_area'], -1) == round(area, -1)), 'rooms'] = median    
    
print('Объявлений без указанного количества комнат после:', df[df['rooms'] == 0]['rooms'].count())
```

Артефакты в столбце с количеством комнат исправлены. Теперь заполним пропуски и значения ниже порогового медианным значением жилой площади по квартирам с аналогичным количеством комнат:

```python
print('Пропуски до:', df['living_area'].isna().sum())

min_living_area = 9  # минимальная регламентированная жилая площадь

for room in df['rooms'].unique(): # для заполнения пропусков используем цикл
    median = df.loc[df['rooms'] == room, 'living_area'].median()  # определяем медиану для каждой комнатности
    df.loc[((df['living_area'].isna()) & (df['rooms'] == room)) |
           ((df['living_area'] < min_living_area) & (df['rooms'] == room)), 'living_area'] = median  # используем метод loc для заполнения пропусков

print('Пропуски после:', df['living_area'].isna().sum())
```

Выбросы и пропуски в столбце `living_area` устранены.

```python
display(pd.DataFrame(df['kitchen_area'].describe()))

df['kitchen_area'].hist(bins=50)
plt.xlabel('Площадь, м²')
plt.ylabel('Количество')
plt.show()
```

Наблюдается распределение с 1 пиком, но есть длинный диапазон из больших значений. Проанализируем данные более детально:

```python
plt.figure(figsize=(10, 5))
ax = sns.scatterplot(data=df, x='total_area', y='kitchen_area')
plt.title('Распределение общей площади и площади кухни', fontsize=14, pad=20)   
plt.xlabel('Общая площадь, м²')
plt.ylabel('Площадь кухни, м²')
sns.despine()
plt.show()
```

На диаграмме рассеяния видна прямая зависимость площади кухни и общей площади. Таким образом мы убедились, что большая площадь кухни обусловлена размером общей площади квартиры. Для заполнения пропусков в площади кухни сначала необходимо определить медианную доли площади кухни к общей площади по квартирам с анлогичным числом комнат. Площадь кухни далее найдем как произведение медианной доли площади кухни на общую площадь. Сначала убедимся, что в столбце `total_area` нет выбросов:

```python
#  проверим объявления, где жилая площадь больше общей 
print('Количество артефактов в столбце "total_area":', df[df['living_area'] > df['total_area']]['total_area'].count())
```

Обнаружено 29 объявлений, где жилая площадь больше общей площади квартиры, чего не может быть по определению. Заменим выбросы на произведение жилой площади и медианного соотношения общей площади к жилой:

```python
# объявим глобальные переменные
median = (df['total_area'] / df['living_area']).median()  
q1 = (df['total_area'] / df['living_area']).quantile(0.25)
q3 = (df['total_area'] / df['living_area']).quantile(0.75)    
iqr = 1.5 * (q3 - q1)

# напишем функцию для проверки вхождения в межквартильный размах
def total_area_adj(row):
    share = row['living_area']/row['total_area']
    living_area = row['living_area']
    total_area = row['total_area']     
    if share > 1:
        return median * living_area
    return total_area

# применим функцию к столбцу
df['total_area'] = df.apply(total_area_adj, axis=1)

print('Количество артефактов в столбце "total_area":', df[(df['living_area']/df['total_area']) > 1]['total_area'].count())
```

Выбросы в столбце `total_area` устранены. Теперь создадим столбец `kitchen_area_%` с данными по доле площади кухни от общей площади и изучим данные более детально:

```python
df['kitchen_area_%'] = df['kitchen_area'] / df['total_area']

display(pd.DataFrame(df['kitchen_area_%'].describe()))
df['kitchen_area_%'].hist(bins=50)
plt.xlabel('Доля площади кухни')
plt.ylabel('Количество')
plt.show()
```

Распределение выглядит логичным, низкая и высокая доля площади кухни характерна для квартир с большой и малой общей площадью соответственно. Заменим пропуски на медианную долю по квартирам с аналогичным числом комнат и расчитаем площадь кухни как произведение `kitchen_area_%` на `total_area`:

```python
print('Пропуски до:', df['kitchen_area'].isna().sum())

for room in df['rooms'].unique(): # для заполнения пропусков используем цикл
    median = df.loc[df['rooms'] == room, 'kitchen_area_%'].median()  # определяем медиану для каждой комнатности   
    df.loc[(df['kitchen_area_%'].isna()) & (df['rooms'] == room), 'kitchen_area_%'] = median  # используем метод loc для заполнения пропусков

df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df['kitchen_area_%'] * df['total_area']
    
print('Пропуски после:', df['kitchen_area'].isna().sum())
```

Изучим данные после заполнения:

После заполнения пропусков на гистограмме более явным стал второй пик доли площади кухни в районе значения 0.25. Эта доля соответствует 1-комнатным квартирам и квартирам со свободной планировкой, исходя из чего, можно сделать вывод, что объявления с отстутствующими данными по площади кухни соответсвтуют квартирам-студиям, где площадь кухни в явном виде не определена.

### Заполнение пропусков в столбце `locality_name`

Т.к. релевантные данные для заполнения пропусков отсутсвуют, заменим пропуски на пустую строку:

```python
print('Пропуски до:', df['locality_name'].isna().sum())
df['locality_name'] = df['locality_name'].fillna('')
print('Пропуски после:', df['locality_name'].isna().sum())
```

Значения в данном столбце являются категориальной признаком объявления, изучим принимаемые им значения:

```python
pd.DataFrame(df['locality_name'].value_counts().head(20))
```

Необходимо привести все названия к единому виду, т.к. из таблицы выше видно, что один и тот же населённый пункт может быть записан различными способами (деревня Кудрово, Кудрово). Для начала определим типы населённых пунктов, встречающиеся в таблице:

```python
# напишем функцию для выделения части строки с нижним регистром
def locality_adj(text):
    return ' '.join([word for word in text.split() if word[0].islower()])

df['locality_type'] = df['locality_name'].apply(locality_adj) # применим функцию к столбцу

locality_type_list = df['locality_type'].drop_duplicates().reset_index(drop=True)  
pd.DataFrame(locality_type_list)
```

Создан новый столбец `locality_type` с типом населённого пункта, теперь можно исключить их из самих названий населённых пунктов:

```python
for word in locality_type_list: # используем цикл для замены типов населённых пунктов на пустую строку
    df['locality_name'] = df['locality_name'].str.replace(word,'')
```

Пропуски устранены, а названия населённых пунктов приведены к единому формату.

### Заполнение пропусков в данных о числе объектов в окрестностнях населённого пункта и расстоянии до них

Пропуски в данных о расстоянии до объектов и их числе этих объектов заполним медианным расстоянием и количеством объектов по анлогичному населённому пункту:

```python
columns = ['airports_nearest','city_centers_nearest','parks_around3000','parks_nearest','ponds_around3000','ponds_nearest']

for column in columns:  # через цикл заполняем данные в столбцах
    
    print('Пропуски в столбце ' + column + ' до:', df[column].isna().sum())

    for word in df['locality_name'].unique():
        median = df.loc[df['locality_name'] == word, column].median()  # определяем медиану для каждого населённого пункта
        df.loc[(df[column].isna()) & (df['locality_name'] == word), column] = median  # используем метод loc для заполнения пропусков

    print('Пропуски в столбце ' + column + ' после:', df[column].isna().sum())    
```

Часть пропусков заполнена, но изначально в этих столбцах было много пропусков. Незаполненные пропуски оставим как есть.

### Заполнение пропусков в `days_exposition` 

В данном столбце содержится информация о количестве дней, прошедшем от размещения объявления до снятия. Необходимо проверить гипотезу о том, что пропуски в данном столбце соответсвуют незакрытым объявлениям. Для начала переведем значения столбца `first_day_exposition` в формат даты:

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Построим график, где по оси X будет отражаться дата размещения объявления, а по Y - количество пропусков в столбце:

```python
df['days_exposition_na'] = df['days_exposition'].isna().astype(int)
report = df.groupby('first_day_exposition')['days_exposition_na'].sum().reset_index()

plt.figure(figsize=(15, 5))
ax = sns.lineplot(x='first_day_exposition', y='days_exposition_na', data=report)
plt.title('Распределение объявлений с пропусками в "days_exposition"', fontsize=16, pad=20) 
plt.xlabel('Дата')
plt.ylabel('Количество') 
plt.show()
```

Как видно из распределения, чем позже было опубликовано объявление, тем больше наблюдается пропусков в данных. Это подтверждает гипотезу о том, что пропуски соответсвуют незакрытым объявлениям. Данный тип пропусков является MNAR, т.к. пропуски обусловлены особенностями порядка сбора информации, поэтому будем заполнять его релевантными данными. В данном случае это будет разница между датой размещения объявления и датой выгрузки таблицы, так мы сможем ранжировать объявления по сроку их жизни на сайте до определенного момента времени.

```python
print('Пропуски до:', df['days_exposition'].isna().sum())

# объявим глобальные переменные
report_date = pd.to_datetime("2021-11-21")

# напишем функцию для нахождения разницы дат
def days_exposition_adj(row):
    exposition_date = row['first_day_exposition']  
    exposition_days = row['days_exposition']
    
    if pd.isna(exposition_days):
        return (report_date - exposition_date).days
    return exposition_days

# применим функцию к столбцу
df['days_exposition'] = df.apply(days_exposition_adj, axis=1)

print('Пропуски после:', df['days_exposition'].isna().sum())
```

Пропуски в столбце устранены.

### Заполнение пропусков в столбце `floors_total`

Пропуски в данном столбце заполним медианной этажностью зданий в аналогичном населённом пункте: 

```python
print('Пропуски до:', df['floors_total'].isna().sum())

for locality in df['locality_name'].unique(): # для заполнения пропусков используем цикл
    median = df.loc[df['locality_name'] == locality, 'floors_total'].median()  # определяем медиану для каждого населённого пункта
    df.loc[(df['floors_total'].isna()) & (df['locality_name'] == locality) 
           & (df['floor'] <= median), 'floors_total'] = median  # используем метод loc для заполнения пропусков, если медиана больше или равна этажу квартиры
df.loc[(df['floors_total'].isna()), 'floors_total'] = df.loc[(df['floors_total'].isna()), 'floor']  # в противном случае заполняем пропуск текущим этажом
    
print('Пропуски после:', df['floors_total'].isna().sum())
```

Пропуски устранены.

### Анализ 10 населённых пунктов с наибольшим числом объявлений

Проанализируем более подробно данные в населённых пунктах с наибольшим числом объявлений:

```python
top_10_list = df['locality_name'].value_counts().head(10)  # список топ-10 населённых пунктов
report = df.query('locality_name in @top_10_list.index')  # формируем срез по вхождению в список

def price_per_sqm(row):  #напишем функцию для определения средней стоимости кв. м. по топ-10
    return row['last_price'].sum() / row['total_area'].sum()

pd.DataFrame(report.groupby('locality_name').apply(price_per_sqm).astype('int').sort_values()
             .reset_index().rename(columns={'locality_name':'населённый пункт', 0:'стоимость 1 м²'}))
```

Самая низкая стоимость 1 м² жилья наблюдается в Выборге и составляет 57 898 руб., а самая высокая в Санкт-Петербурге, где средняя стоимость 1 м² равняется 124 401 руб.
